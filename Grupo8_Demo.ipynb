```python
#  Definir columnas del modelo
VARS_NUM = ['duracion_min_trans', 'tempo', 'bailabilidad', 'energia', 'volumen',
            'tonalidad', 'positividad', 'acustica_trans', 'discursividad_trans',
            'vitalidad_trans']

COLUMNAS = ['duracion_min_trans', 'tempo', 'bailabilidad', 'energia', 'volumen',
            'tonalidad', 'positividad', 'acustica_trans', 'discursividad_trans',
            'vitalidad_trans', 'categoria_antiguedad_Media',
            'categoria_antiguedad_Reciente', 'region_LatAm',
            'region_Asia_Pacifico', 'region_Otro', 'region_MENA',
            'region_Europa', 'region_Africa', 'genero_cluster_1',
            'genero_cluster_2', 'genero_cluster_3', 'genero_cluster_4',
            'modalidad_armonica_1']

print(f"Variables del modelo: {len(COLUMNAS)}")
```

```python
import pandas as pd

# Función de predicción
def predecir_popularidad(bailabilidad, energia, volumen, tempo, positividad,
                          acustica_trans, discursividad_trans, vitalidad_trans,
                          duracion_min_trans, tonalidad, modalidad_mayor,
                          region, antiguedad, genero_cluster):
    """
    Predice la popularidad de una canción.

    Parámetros de audio (0-1 excepto donde se indica):
    - bailabilidad: qué tan bailable es (0-1)
    - energia: intensidad y actividad (0-1)
    - volumen: volumen en dB (típico: -15 a 0)
    - tempo: BPM (típico: 60-200)
    - positividad: tono emocional positivo (0-1)
    - acustica_trans: nivel acústico transformado
    - discursividad_trans: presencia de voz hablada transformada
    - vitalidad_trans: presencia de audiencia en vivo transformada
    - duracion_min_trans: duración transformada
    - tonalidad: 0-11 (Do=0, Do#=1, Re=2...)
    - modalidad_mayor: 1=Mayor, 0=Menor

    Parámetros categóricos:
    - region: 'LatAm', 'Europa', 'Asia_Pacifico', 'MENA', 'Africa',
              'Norte_America', 'Otro'
    - antiguedad: 'Reciente', 'Media', 'Antigua'
    - genero_cluster: 0, 1, 2, 3, 4
    """

    datos = {col: 0 for col in COLUMNAS}

    # Variables numéricas
    datos['bailabilidad'] = bailabilidad
    datos['energia'] = energia
    datos['volumen'] = volumen
    datos['tempo'] = tempo
    datos['positividad'] = positividad
    datos['acustica_trans'] = acustica_trans
    datos['discursividad_trans'] = discursividad_trans
    datos['vitalidad_trans'] = vitalidad_trans
    datos['duracion_min_trans'] = duracion_min_trans
    datos['tonalidad'] = tonalidad
    datos['modalidad_armonica_1'] = modalidad_mayor

    # Region (one-hot)
    regiones_map = {
        'LatAm': 'region_LatAm',
        'Europa': 'region_Europa',
        'Asia_Pacifico': 'region_Asia_Pacifico',
        'MENA': 'region_MENA',
        'Africa': 'region_Africa',
        'Otro': 'region_Otro'
    }
    if region in regiones_map:
        datos[regiones_map[region]] = 1

    # Antiguedad (one-hot)
    if antiguedad == 'Media':
        datos['categoria_antiguedad_Media'] = 1
    elif antiguedad == 'Reciente':
        datos['categoria_antiguedad_Reciente'] = 1

    # Genero cluster (one-hot)
    if genero_cluster in [1, 2, 3, 4]:
        datos[f'genero_cluster_{genero_cluster}'] = 1

    df_input = pd.DataFrame([datos])
    prediccion = modelo.predict(df_input)[0]

    return round(prediccion, 2)

print("✓ Función de predicción lista")
```

```python
#  Ejemplo 1 - Pop/Dance en Latinoamérica ---

print("EJEMPLO 1: Canción Pop/Dance en Latinoamérica")
print("=" * 50)
pred_1 = predecir_popularidad(
    bailabilidad=0.75,
    energia=0.80,
    volumen=-4.5,
    tempo=120,
    positividad=0.65,
    acustica_trans=0.15,
    discursividad_trans=0.08,
    vitalidad_trans=0.12,
    duracion_min_trans=0.98,
    tonalidad=1,
    modalidad_mayor=1,
    region='LatAm',
    antiguedad='Reciente',
    genero_cluster=1
)
print(f"Popularidad predicha: {pred_1}")
```
