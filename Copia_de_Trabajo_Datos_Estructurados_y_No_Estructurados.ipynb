```python
import json

mcc = pd.read_json('/content/mcc_codes.json', orient='index')
mcc.reset_index(inplace=True)
mcc.columns = ['MCC_Code', 'Description']
mcc.head()
```

# **3.  Limpieza de Datos**

---

```python
# Convertir fechas a datetime
cards["acct_open_date"] = pd.to_datetime(cards["acct_open_date"], format='%b-%y', errors='coerce')
cards["expires"] = pd.to_datetime(cards["expires"], format='%b-%y', errors='coerce')
tx["date"] = pd.to_datetime(tx["date"], format='%d/%m/%Y %H:%M', errors='coerce')

print("Cards DataFrame head:")
print(cards.head())
print("\nTransactions DataFrame head:")
print(tx.head())
```

```python
# Quitar espacios en columnas categóricas para DataFrames
for df_obj in [cards, tx, users]:
    df_obj.columns = df_obj.columns.str.strip()

print("Cards DataFrame head (adjusted columns):")
print(cards.head())
print("\nTransactions DataFrame head (adjusted columns):")
print(tx.head())
print("\nUsers DataFrame head (adjusted columns):")
print(users.head())
```

```python
# Limpiar y convertir 'yearly_income' a numérico
users['yearly_income'] = users['yearly_income'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)

# Limpiar y convertir 'total_debt' a numérico
users['total_debt'] = users['total_debt'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)

users.head()
```

```python
# Transactions ↔ MCC
tx_mcc = tx.merge(mcc, left_on="mcc", right_on="MCC_Code", how="left")

tx_mcc.head()
```

```python
# Users + Cards + Transactions + MCC
full = (users
        .merge(cards, left_on="id", right_on="client_id", how="left")
        .merge(tx, left_on="id_y", right_on="card_id", how="left")
        .merge(mcc, left_on="mcc", right_on="MCC_Code", how="left")
       )

full.head()
```

```python
# === Indicadores Transactions_data ===

tx["amount"] = tx["amount"].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

```python
# ----- Total transacciones -----
plt.figure()
plt.hist(tx["amount"])
plt.title("Distribución del valor de transacciones")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# ----------  Indicadores Avanzados  --------------

# 1. Gasto mensual por usuario

tx["month"] = tx["date"].dt.to_period("M")
gasto_mensual = tx.groupby("month")["amount"].sum()

plt.figure()
gasto_mensual.plot(kind="line")
plt.title("Gasto total por mes")
plt.xlabel("Mes")
plt.ylabel("Monto total")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Gasto por categoría MCC

# Agrupación y suma del gasto por categoría MCC
gasto_mcc = tx_mcc.groupby("Description")["amount"].sum().sort_values(ascending=False).head(10)

# Gráfico
gasto_mcc = pd.merge(tx, mcc, left_on='mcc', right_on='MCC_Code', how='left')
mcc_description_counts = gasto_mcc['Description'].value_counts().nlargest(20)
top_10_mcc_description_percentage = (mcc_description_counts / len(gasto_mcc)) * 100

plt.figure(figsize=(12,6))
ax = sns.barplot(x=top_10_mcc_description_percentage.index, y=top_10_mcc_description_percentage.values, palette="viridis")
plt.xlabel("MCC Description")
plt.title("Top 20 Actividades Economicas con Mayor número de Transacciones")
plt.xticks(rotation=45, ha='right')

# Incluye los porcentajes
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
# INDICADOR: Gasto por categoría MCC por género

# Merge con usuarios (para traer género)
tx_full = tx_mcc.merge(users[["id", "gender"]],
                       left_on="client_id",
                       right_on="id",
                       how="left")

tx_full['amount'] = tx_full['amount'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)

# AGRUPACIÓN: Gasto total por género y MCC

gasto_mcc_genero = (
    tx_full.groupby(["gender", "Description"])["amount"]
           .sum()
           .reset_index())

# Tomar el TOP 10 MCC para cada género

top_mcc_genero = (gasto_mcc_genero
    .sort_values(["gender", "amount"], ascending=[True, False])
    .groupby("gender")
    .head(10))

# GRÁFICO

plt.figure(figsize=(14, 6))

# Pivot para usar en gráfico
pivot_data = top_mcc_genero.pivot(index="Description", columns="gender", values="amount")

pivot_data.plot(kind="bar", figsize=(14,6))

plt.title("Gasto por categoría MCC segmentado por género")
plt.xlabel("Categoría MCC")
plt.ylabel("Gasto total")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# 4. Gasto por ciudad
ciudades = tx.groupby("merchant_city")["amount"].sum().sort_values(ascending=False).head(10)

plt.figure()
plt.bar(ciudades.index.astype(str), ciudades.values)
plt.title("Top 10 ciudades por monto transado")
plt.ylabel("Monto total")
plt.xticks(rotation=90)
plt.show()
```

```python
# --- JOIN ENTRE USERS, CARDS Y TRANSACTIONS ---

# First, merge transactions with user information
tx_with_users = tx.merge(users, left_on='client_id', right_on='id', how='left', suffixes=('_tx', '_user'))

# Prepare cards dataframe for merge, explicitly selecting and renaming card_type
cards_for_merge = cards[['id', 'card_type']].rename(columns={'id': 'card_id_from_cards_df'})

# Then, merge the result with card information
# Use card_id from tx_with_users and id from original cards df (now card_id_from_cards_df)
us_ca_tx = tx_with_users.merge(cards_for_merge, left_on='card_id', right_on='card_id_from_cards_df', how='left')

# Drop the temporary merge key column if it's not needed later
us_ca_tx.drop(columns=['card_id_from_cards_df'], inplace=True)

# Now, 'card_type' should exist directly in us_ca_tx without any suffix

# --- INDICADOR: CANTIDAD DE USUARIOS POR MERCHANT CITY Y TIPO DE TARJETA ---

# Agrupar contando usuarios únicos por ciudad y tipo de tarjeta
city_card_users = us_ca_tx.dropna(subset=['merchant_city', 'card_type']).groupby(["merchant_city", "card_type"])["client_id"].nunique().reset_index()

# Renombrar columna
city_card_users.rename(columns={
    "client_id": "cantidad_usuarios"
}, inplace=True)

# --- TOP 10 CIUDADES POR CANTIDAD TOTAL DE USUARIOS ---
top10_cities = city_card_users.groupby("merchant_city")["cantidad_usuarios"].sum() \
                              .sort_values(ascending=False) \
                              .head(10) \
                              .index

city_card_users_top10 = city_card_users[city_card_users["merchant_city"].isin(top10_cities)]

# --- GRÁFICO DE BARRAS APILADAS ---

import matplotlib.pyplot as plt
import matplotlib.cm as cm # Import colormap
import numpy as np # Import numpy

plt.figure(figsize=(14, 7))

pivot_df = city_card_users_top10.pivot(index="merchant_city",
                                       columns="card_type",
                                       values="cantidad_usuarios")

# Define custom colors for the card types dynamically
num_categories = len(pivot_df.columns)
colors = cm.viridis(np.linspace(0, 1, num_categories))

ax = pivot_df.plot(kind="bar", stacked=True, figsize=(14, 7), color=colors)

plt.title("Cantidad de usuarios por ciudad y tipo de tarjeta (Top 10 ciudades)")
plt.xlabel("Ciudad")
plt.ylabel("Cantidad de usuarios")
plt.xticks(rotation=45)
plt.legend(title="Tipo de tarjeta")

plt.tight_layout()
plt.show()
```

```python
# AGRUPAR MONTO TOTAL POR CATEGORÍA DE CONSUMO Y TIPO DE TARJETA =====

# Merge tx (transactions) with mcc (merchant category codes) to get the 'Description'
tx_with_mcc = tx.merge(mcc, left_on='mcc', right_on='MCC_Code', how='left')

# Merge the result with cards to get the 'card_brand'
# We select only 'id' and 'card_brand' from cards to merge, to avoid duplicate columns
transactions_full = tx_with_mcc.merge(cards[['id', 'card_brand']], left_on='card_id', right_on='id', how='left', suffixes=('_tx', '_card'))

# Filter for only Visa and Mastercard transactions
mcc_type_amount = transactions_full[transactions_full["card_brand"].isin(["Visa", "Mastercard"])]

# Group by consumption 'Description' and 'card_brand', then sum the 'amount'
mcc_type_amount_grouped = mcc_type_amount.groupby(["Description", "card_brand"])["amount"].sum().reset_index()

# Calculate total amount per MCC Description to identify the top 10
total_amount_per_mcc = mcc_type_amount_grouped.groupby("Description")["amount"].sum().sort_values(ascending=False)
top_10_mcc_descriptions = total_amount_per_mcc.head(10).index

# Filter the grouped data for only the top 10 MCC descriptions
mcc_type_amount_grouped_top10 = mcc_type_amount_grouped[mcc_type_amount_grouped["Description"].isin(top_10_mcc_descriptions)]

# Pivot the table to prepare for plotting, with 'Description' as index and 'card_brand' as columns
pivot_df = mcc_type_amount_grouped_top10.pivot(index="Description", columns="card_brand", values="amount").fillna(0)

# ===== GRÁFICO: Monto total Visa vs Mastercard por categoría de consumo (Top 10 MCC) =====

plt.figure(figsize=(15, 7))

# Define custom colors for Visa and Mastercard
colors = cm.viridis(np.linspace(0, 1, num_categories))

pivot_df.plot(kind="bar", figsize=(15, 7), color=colors)

plt.title("Monto total de transacciones Visa vs Mastercard por categoría de consumo (Top 10 MCC)")
plt.xlabel("Categoría de consumo (MCC)")
plt.ylabel("Monto total de transacciones")
plt.xticks(rotation=70)
plt.legend(title="Tipo de Tarjeta")
plt.tight_layout()
plt.show()
```
